### Una vez cargados todos los datasets estamos listos para comenzamos con la Transformación de los Datos 

Amazon

```python
#Creo el campo 'ID':
df_amazon['ID'] = ['a'] + df_amazon['show_id']


#Remplazo los valores nulos en la columna Rating por 'G':
df_amazon['rating'] = df_amazon['rating'].replace(np.nan, 'G')

#Normalizo formato de fechas
df_amazon['date_added'] = pd.to_datetime(df_amazon['date_added'], format = '%B %d, %Y')

#Cambio los campos de txt para que esten en minuscula:
df_amazon['type'] = df_amazon['type'].str.lower()
df_amazon['title'] = df_amazon['title'].str.lower()
df_amazon['director'] = df_amazon['director'].str.lower()
df_amazon['cast'] = df_amazon['cast'].str.lower()
df_amazon['country'] = df_amazon['country'].str.lower()
df_amazon['rating'] = df_amazon['rating'].str.lower()
df_amazon['duration'] = df_amazon['duration'].str.lower()
df_amazon['listed_in'] = df_amazon['listed_in'].str.lower()
df_amazon['description'] = df_amazon['description'].str.lower()
```

```python
#Dividido el campo 'duratiom' en dos: 'duration_int' y 'duration_type':
df_amazon[['duration_int', 'duration_type']] = df_amazon['duration'].str.extract('(\d+)\s*(\w+)')
```

```python
#Reviso que todos los cambios se hayan aplicado correctamente
df_amazon
```

```python
#Remplazo el valor 'seasons' por 'season' en el campo 'duration_type'
df_amazon['duration_type'] = df_amazon['duration_type'].str.replace('seasons', 'season')
```

```python
#Vuelvo a revisar que los cambios se hayan realizado correctamente
df_amazon
```

### Creamos un solo dataset con los 4 datasets

```python
#Primero voy a agregar a todos los datasts una columna llamada 'platform'
df_amazon['platform'] = 'amazon'
df_disney['platform'] = 'disney'
df_hulu['platform'] = 'hulu'
df_netflix['platform'] = 'netflix'
```
